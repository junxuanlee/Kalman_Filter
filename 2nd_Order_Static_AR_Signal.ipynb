```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from functions_KF import * 
```

# Generate 2nd order static AR signal

```python
# Length of time series
N=400
A = np.array([1.5, -0.9])

signal_model = static_2nd_AR(N,A)
signal_model.generate_signal()

S = signal_model.S
ex = signal_model.ex

signal_model.plot()
```

# Generate parameters

```python
N=400

theta_n1_n1 = np.random.randn(2,1) # can change to np.array([[1],[2]]) or np.random.randn(2,1)

P_n1_n1 =  0.001*np.identity(2) #larger value gives faster convergence

# Initial data/sample value
x_n = np.zeros((2,1))

# Process noise
Q = 0.001*np.identity(2)

# Measurement noise
R = 0.2*np.cov(ex) ## std
```

# Run Kalman Filter

```python
KF = kalman_filter(N, S, theta_n1_n1, P_n1_n1, x_n, Q, R,A,'static')
KF.train()
KF.plot()
```
